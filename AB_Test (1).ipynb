```python
# user reactions occured
newdf = df[(df.yes == 1) | (df.no == 1)]
newdf.head()
```

```python
newdf.count()
```

```python
newdf_exposed = newdf[(newdf.experiment == 'exposed')]
newdf_control = newdf[(newdf.experiment == 'control')]
```

```python
newdf_exposed.count()
```

```python
newdf_control.count()
```

```python
newdf_control.head()
```

```python
from statsmodels.stats.proportion import proportions_ztest, proportion_confint

c_yes = newdf_control['yes']
e_yes = newdf_exposed['yes']

n_con = c_yes.count()
n_exposed = e_yes.count()
successes = [c_yes.sum(), e_yes.sum()]
nobs = [n_con, n_exposed]
```

```python
z_stat, pval = proportions_ztest(successes, nobs=nobs)
(lower_con, lower_treat), (upper_con, upper_treat) = proportion_confint(successes, nobs=nobs, alpha=0.05)

print(f'z statistic: {z_stat:.2f}')
print(f'p-value: {pval:.3f}')
print(f'ci 95% for control group: [{lower_con:.3f}, {upper_con:.3f}]')
print(f'ci 95% for exposed group: [{lower_treat:.3f}, {upper_treat:.3f}]')
```

p-value=0.732 is way above our α=0.05 so we can not reject null hypothesis that no significant differance between two cases
